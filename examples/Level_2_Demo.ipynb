## üéØ Selection Operators Demo

Selection operators are genome-agnostic and work purely on fitness values. Let's demonstrate **Tournament** and **Roulette Wheel** selection.

````python
# üéØ Selection Operators Demo

# Create a population with fitness values (higher = better)
population_size = 20
key_select = jar.split(key10)[0]  # Use key10 from categorical demo
fitness_values = jar.uniform(key_select, (population_size,), minval=0.0, maxval=100.0)

print("üéØ Selection Operators Demo:")
print(f"Population size: {population_size}")
print(f"Fitness values: {np.round(fitness_values, 1)}")
print(f"Best fitness: {np.max(fitness_values):.1f} (index {np.argmax(fitness_values)})")
print(f"Worst fitness: {np.min(fitness_values):.1f} (index {np.argmin(fitness_values)})")

# Tournament Selection (New Paradigm)
tournament_selector = TournamentSelection(num_selections=5, tournament_size=4)

key_tour = jar.split(key_select)[0]
selected_indices = tournament_selector(key_tour, fitness_values)

print(f"\nüèÜ Tournament Selection Results (5 selections, tournament_size=4):")
print(f"Selected indices: {selected_indices}")
print(f"Selected fitness: {np.round(fitness_values[selected_indices], 1)}")

# Roulette Wheel Selection (New Paradigm)
roulette_selector = RouletteWheelSelection(num_selections=6)

key_roul = jar.split(key_tour)[0]
roulette_indices = roulette_selector(key_roul, fitness_values)

print(f"\nüé∞ Roulette Wheel Selection Results (6 selections):")
print(f"Selected indices: {roulette_indices}")
print(f"Selected fitness: {np.round(fitness_values[roulette_indices], 1)}")

# Selection pressure comparison
print(f"\nüìä Selection Pressure Analysis:")
tournament_avg = np.mean(fitness_values[selected_indices])
roulette_avg = np.mean(fitness_values[roulette_indices])
population_avg = np.mean(fitness_values)

print(f"  Population average:  {population_avg:.1f}")
print(f"  Tournament average:  {tournament_avg:.1f} (pressure: {tournament_avg/population_avg:.2f}x)")
print(f"  Roulette average:    {roulette_avg:.1f} (pressure: {roulette_avg/population_avg:.2f}x)")

# Test JIT compilation
print(f"\n‚ö° Testing JIT Compilation:")
jit_tournament = jax.jit(tournament_selector)
jit_roulette = jax.jit(roulette_selector)

jit_tournament_result = jit_tournament(jar.PRNGKey(999), fitness_values)
jit_roulette_result = jit_roulette(jar.PRNGKey(888), fitness_values)

print(f"  ‚úì JIT Tournament result: {jit_tournament_result}")
print(f"  ‚úì JIT Roulette result: {jit_roulette_result}")

print(f"\n‚úì Tournament selection provides higher selection pressure!")
print(f"‚úì Roulette wheel maintains fitness-proportional selection!")
print(f"‚úì Both operators use new @struct.dataclass paradigm!")
print(f"‚úì Full JIT compilation support with static parameters!")
````
